```python
features = ["keywords","genres","cast","director"]
for i in features:
    file[i] = file[i].fillna(" ")
```

### Adding new column which contrains string of all features

```python
def combined_features(row):
    return row["keywords"] + " " + row["genres"] + " " + row["cast"] + " " + row["director"] 
```

```python
file["combined_features"] = file.apply(combined_features, axis=1)
```

### Computing the cosine similarity after vectorizing features

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(file["combined_features"])
#print(count_matrix) #it can also be converted to numpy array by count_matrix.toarray()
```

```python
cos_sim = cosine_similarity(count_matrix)
print(cos_sim)
```

```python
def get_index(movie):
    return file[file["title"] == movie]["index"].values[0]
def get_movie(index):
    return file[file["index"] == index]["title"].values[0]
```

```python
movie_index = get_index('Bad Moms')
similiar_movies = list(enumerate(cos_sim[movie_index]))
```

```python
sorted_movies = sorted(similiar_movies,key=lambda x:x[1],reverse=True)
sorted_movies
```

```python
top_matched = []
for i in sorted_movies:
    if i[1] > 0.1:
        top_matched.append(i[0])
        
```

```python
recommended_movies = []
for i in top_matched:
    recommended_movies.append(get_movie(i))
```

```python
for i in recommended_movies:
    print(i)
    print("----")
```

```python
def recommend_movies(movie):
    movie_index = get_index(movie)
    similiar_movies = list(enumerate(cos_sim[movie_index]))
    sorted_movies = sorted(similiar_movies,key=lambda x:x[1],reverse=True)
    
    top_matched = []
    
    for i in sorted_movies[1:]:
        if i[1] > 0.1:
            top_matched.append(i[0])
    
    recommended_movies = []
    for i in top_matched:
        recommended_movies.append(get_movie(i))
    
    for i in recommended_movies:
        print(i)
        print("----")
```

```python
def func():
    movie = input("Enter your favourite movie:")
    print("\n")
    print("Because you like " + movie)
    print("\n")
    recommend_movies(movie)
```
